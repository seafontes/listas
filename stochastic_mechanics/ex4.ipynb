```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

problem modeling:
states 0,1,2 where the K_x'x are all one except the 0 <-> 1:
k_10 = exp(delta/2kbT); k_01 = exp(-delta/2kbT)

```python
index = [0,1,2]

delta = 0.001 # due to the bellow, [delta] = [eV] 
const = 2*293*8.61733*(10**-5) # [const] = [][eVK^-1][K]=[eV]

kminus = math.exp(-1*delta/const)
kplus  = math.exp(delta/const)

K = [
    [0, kminus, 1], 
    [kplus,  0, 1],
    [1,     1,  0]
]
```

starting the gillespie algorithm

```python
t_i = 0.0  #starting time
t_f = 20 #finishing time

state = 0
entropy = 0

time = [t_i]
state_t = [state]
entropy_array = [entropy]

t = t_i

while t < t_f:
    #calculate the probabilities:
    minus = (state-1) if (state-1) > -1 else 2
    plus = (state+1)%3
    p_minus = K[minus][state]
    p_plus  = K[plus][state]
    p_0 = p_minus+p_plus

    r1 = np.random.rand()
    r2 = np.random.rand()

    tau = (1/p_0)*math.log(1/r1)

    if r2 < p_minus/p_0:
        state = minus
        entropy += 293*8.61733*(10**-5)*math.log(/)
    else:
        state = plus

    t += tau

    state_t.append(state)
    time.append(t)
    entropy_array.append(entropy)


plt.step(time, state_t, label='State')
plt.xlabel('Time')
plt.ylabel('Count')
plt.legend()
plt.show()
```
